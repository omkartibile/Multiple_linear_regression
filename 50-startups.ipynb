# ****Multiple Linear Regression****

**Basic assumptions of regression**

1.Linearity:- There should be linear relationship between dependent and independent variable

2.Normality:- For any fixed value of x,y should be normally distributed

3.Independency:- All the observations should be independent of each other

4.Homosecdasticity:-Errors should be independent and identically distributed

**Null and alternative hypothesis for regression**

**H0**: There is no relationship between x and y
        i.e. b1=b2=b3....bn=0
        
**H1**: The relation between x and y is not 0
        i.e. at least one of bj not equal 0

```python
#seprating the independent varibales from data
independent=df.drop('Profit',axis=1)
```

```python
#Seprating the dependent varibale
dependent=df['Profit']
```

```python
#Converting the categorical variables to binary format
state=pd.get_dummies(independent,drop_first=True)
```

```python
#removing the categorical variable
independent=independent.drop('State',axis=1)
```

```python
#Concatenating the binary format of categorical variable to the dataframe of independent variables
independent=pd.concat([independent,state],axis=1)
```

```python
#Checking the independent variable dataset
independent.head(2)
```

```python
#Split the data using train test method
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.22,random_state=42)
```

```python
#Checking the size of each splitted dataset
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
#Importing Linear Regression model
from sklearn.linear_model import LinearRegression
```

```python
LR=LinearRegression()
```

```python
#Fitting the model with training data
LR.fit(x_train,y_train)
```

```python
#Predicting the values
y_pred=LR.predict(x_test)
```

```python
y_pred
```

```python
#Importing the evaluation metrics
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
#Analyzing the metrics
Rsquare=r2_score(y_test,y_pred)
print('The R-Square value of model is:', Rsquare)
```

```python
MAE=mean_absolute_error(y_test,y_pred)
print('The MAE value of model is:', MAE)
```

```python
MSE=mean_squared_error(y_test,y_pred)
print('The MSE value of model is:', MSE)
```

# Result
**As we can see we have got the r-square value of 89% we can conclude that our fitted regression model is good for future predictions**
